# Plotting distribution of $\theta$ from the Empirical Bayesian Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def make_plot(mydat, title, fname):
    """ Plotting function for making the various panels of plots below """
    fig, axes = plt.subplots(nrows=4, ncols=2)
    fig.set_size_inches(12,10)
    fig.suptitle(title, fontsize=14)
    
    plot_titles = ('apn < 5', '5 <= apn < 15', '15 <= apn < 50', '50 <= apn')
    all_axes = ((0,0),(0,1),(1,0),(1,1),(2,0),(2,1),(3,0),(3,1))
    labels = ('A', 'B', 'C', 'D', 'E', 'F', 'G', 'H')
    
    for i, dat in enumerate(mydat):
        # Get the current panel of the plot
        ax = axes[all_axes[i]]
        
        # Get the current panel's title, the first 4 plots have bias, the last 4 have no bias
        if i <= 3:
            curr_title = plot_titles[i] + "\nqsim has bias"
        else:
            curr_title = plot_titles[i-4] + "\nqsim has no bias"
            
        dat[['q4', 'q5', 'q6']].plot(ax=ax, kind='density', xlim=(-0.1, 1.1), color=('r','b','k')); 
        ax.set_xlabel('Line <--- AB ---> Tester')
        ax.axvline(x=0.5,linewidth=2, color='k')
        ax.set_title(curr_title)
        ax.text(-0.07, 1.01, labels[i], transform=ax.transAxes,fontsize=16, fontweight='bold', va='top', ha='right')

    plt.tight_layout(h_pad=2,w_pad=2)
    fig.savefig(fname)
```

##Import Empirical Bayesian dataset with flags

```python
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/emp_bayesian/output/emp_for_plotting.csv'
mydat = pd.read_csv(fname)
mydat[:3]
```

## Create subsets
Initial subsetting is done by APN

* r0 is an APN < 5
* r1 is an APN between 5 and 15
* r2 is an APN between 15 and 50
* r3 is an APN > 50

Mated and Virgin is also split for plotting.

Then split if there was bias in qsim.

```python
# Split by APN
r0 = mydat[mydat['rank_apn'] == 0]
r1 = mydat[mydat['rank_apn'] == 1]
r2 = mydat[mydat['rank_apn'] == 2]
r3 = mydat[mydat['rank_apn'] == 3]
```

```python
# Split mated
m0 = r0[r0['ms'].str.contains('M')]
m1 = r1[r1['ms'].str.contains('M')]
m2 = r2[r2['ms'].str.contains('M')]
m3 = r3[r3['ms'].str.contains('M')]
```

```python
# Split virgin
v0 = r0[r0['ms'].str.contains('V')]
v1 = r1[r1['ms'].str.contains('V')]
v2 = r2[r2['ms'].str.contains('V')]
v3 = r3[r3['ms'].str.contains('V')]
```

```python
# Keep fusions that are not flagged as having bias
def drop_bias(mydat, flag=0):
    return mydat[mydat['flag_qsim_bias'] == flag]

mated_bias = [drop_bias(x,flag=1) for x in [m0, m1, m2, m3]]
mated_nobias = [drop_bias(x) for x in [m0, m1, m2, m3]]
mated = mated_bias + mated_nobias

virgin_bias = [drop_bias(x,flag=1) for x in [v0, v1, v2, v3]]
virgin_nobias = [drop_bias(x,flag=1) for x in [v0, v1, v2, v3]]
virgin = virgin_bias + virgin_nobias
```

```python
make_plot(mated, 'Mated', '/home/jfear/mclab/cegs_ase_paper/pipeline_output/emp_bayesian/output/emp_bayesian_density_by_apn_mated_8panel.png')
```

```python
make_plot(virgin, 'Virgin', '/home/jfear/mclab/cegs_ase_paper/pipeline_output/emp_bayesian/output/emp_bayesian_density_by_apn_virgin_8panel.png')
```
