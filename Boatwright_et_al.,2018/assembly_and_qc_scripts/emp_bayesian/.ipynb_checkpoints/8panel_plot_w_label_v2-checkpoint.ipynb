# Plotting distribution of $\theta$ from the Empirical Bayesian Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plot_titles = ('apn < 5', '5 <= apn < 15', '15 <= apn < 50', '50 <= apn')
labels = ('A', 'B', 'C', 'D', 'E', 'F', 'G', 'H')
all_axes = ((0,0),(0,1),(1,0),(1,1),(2,0),(2,1),(3,0),(3,1))
```

```python
def emp_plot(mydat, ax, plot_title, label):
    """ Plotting function for making the plots of empirical values """
    curr_title = plot_title + "\nqsim = 0.5"
    mydat[['q4_mean_theta', 'q5_mean_theta', 'q6_mean_theta']].plot(ax=ax, kind='density', xlim=(-0.1, 1.1), color=('r','b','k')); 
    ax.set_xlabel('Line <--- AB ---> Tester')
    ax.axvline(x=0.5,linewidth=2, color='k')
    ax.set_title(curr_title)
    ax.text(-0.07, 1.01, label, transform=ax.transAxes,fontsize=16, fontweight='bold', va='top', ha='right')
```

```python
def qsim_plot(mydat, ax, plot_title, label):
    """ Plotting function for making the plots of qsim values """
    curr_title = plot_title + u"\nqsim ≠ 0.5"
    mydat['qsim_mean_theta'].plot(ax=ax, kind='density', xlim=(-0.1, 1.1), color='k') 
    ax.set_xlabel('Line <--- AB ---> Tester')
    ax.axvline(x=0.5,linewidth=2, color='k')
    ax.set_title(curr_title)
    ax.text(-0.07, 1.01, label, transform=ax.transAxes,fontsize=16, fontweight='bold', va='top', ha='right')
```

```python
# Keep fusions that are not flagged as having bias
def drop_bias(mydat):
    """ Return fusions that have qsim = 0.5 """
    return mydat[mydat['qsim_line'] == 0.5]

def keep_bias(mydat):
    """ Return fusions that have qsim != 0.5 """
    return mydat[(mydat['qsim_line'] != 0.5) & (mydat['qsim_line'] != 1) & (mydat['qsim_line'] != 0)]
```

##Import Empirical Bayesian dataset with flags

```python
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/ase_results/qsim_emp_theta_w_flag.csv'
mydat = pd.read_csv(fname)
mydat[:3]
```

## Create subsets
Initial subsetting is done by APN

* r0 is an APN < 5
* r1 is an APN between 5 and 15
* r2 is an APN between 15 and 50
* r3 is an APN > 50

Mated and Virgin is also split for plotting.

Then split if there was bias in qsim.

```python
# Split by APN
r0 = mydat[mydat['mean_apn'] < 5]
r1 = mydat[(mydat['mean_apn'] >= 5) & (mydat['mean_apn'] < 15)]
r2 = mydat[(mydat['mean_apn'] >= 15) & (mydat['mean_apn'] < 50)]
r3 = mydat[mydat['mean_apn'] > 50]
```

## Mated

```python
# Split mated
m0 = r0[r0['mating_status'].str.contains('M')]
m1 = r1[r1['mating_status'].str.contains('M')]
m2 = r2[r2['mating_status'].str.contains('M')]
m3 = r3[r3['mating_status'].str.contains('M')]

mated_bias = [keep_bias(x) for x in [m0, m1, m2, m3]]
mated_nobias = [drop_bias(x) for x in [m0, m1, m2, m3]]
mated = mated_nobias + mated_bias
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2)
fig.set_size_inches(12,10)
for i, dat in enumerate(mated):
    if i <= 3:
        emp_plot(dat, axes[all_axes[i]], plot_titles[i], labels[i])
    else:
        qsim_plot(dat, axes[all_axes[i]], plot_titles[i-4], labels[i])
fig.suptitle('Mated')
plt.tight_layout(h_pad=2, w_pad=2)
fig.savefig('/home/jfear/mclab/cegs_ase_paper/pipeline_output/emp_bayesian/output/emp_bayesian_density_by_apn_mated_8panel.png')
```

## Virgin

```python
# Split virgin
v0 = r0[r0['mating_status'].str.contains('V')]
v1 = r1[r1['mating_status'].str.contains('V')]
v2 = r2[r2['mating_status'].str.contains('V')]
v3 = r3[r3['mating_status'].str.contains('V')]

virgin_bias = [keep_bias(x) for x in [v0, v1, v2, v3]]
virgin_nobias = [drop_bias(x) for x in [v0, v1, v2, v3]]
virgin = virgin_nobias + virgin_bias
```
