```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/emp_bayesian/output/MBE_PG_emp_bayesian_results.csv'
mydat = pd.read_csv(fname)
mydat[:1]
```

```python
there = mydat[np.isfinite(mydat['q4_mean_theta'])]
qs = (1, 3, 6, 9) # the column number of the mean thatas
df = there.ix[:,qs]
```

```python
ax = df.plot(kind='density', figsize=(10,6), color=('r','b','k'), xlim=(-0.1, 1.1))
ax.axvline(x=0.5, color='k', linewidth=2)
ax.set_xlabel('Line <---- AB ----> Tester')
ax.set_title('Updated with Indels')

fig = ax.get_figure()
fig.savefig('/home/jfear/mclab/cegs_ase_paper/pipeline_output/emp_bayesian/output/emp_bayesian_density.png')
```
